### Overdamped response

```python
import numpy as np
import control as co
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
s=co.tf("s")
G=9/(s**2+9*s+9)
C=9/(s*(s**2+9*s+9))
poles=np.around(C.pole(), 2)
poles
```

```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].plot(poles.real, poles.imag, 
           marker="x", 
           color="red", 
           linestyle="none", 
           markersize=15)
def spine_origin(axes_index, xlabel, ylabel, title):
    #move left y axis and bottom x axis to center (0,0)
    ax[axes_index].spines["left"].set_position(("data",0))
    ax[axes_index].spines["bottom"].set_position(("data", 0))
    #Eliminate upper and right axes
    ax[axes_index].spines["right"].set_visible(False)
    ax[axes_index].spines["top"].set_visible(False)
    ax[axes_index].set_xlabel(xlabel)
    ax[axes_index].set_ylabel(ylabel)
    ax[axes_index].set_title(title, fontdict=font)

# defint text font properties
font = {'family': 'serif',
        'color':  'DarkGreen',
        'weight': 'normal',
        'size': 18,
        }
spine_origin(0, None, None, "Poles plot on S plane")
ax[0].grid()

#Step Response
t, y = co.step_response(G) # calculate the step response of G1
ax[1].plot(t, y, color="red")
spine_origin(1, "time", "Amplitude", "Overdamped Response")
ax[1].grid()
```

```python
zeta=np.arange(0.1, 0.9, 0.1)
omega_t=np.arange(0, 17, 0.1)
fig, ax=plt.subplots(figsize=(15,7))
for zeta in zeta:
    #angle beta calculation
    angle=np.arctan(np.sqrt(1-zeta**2)/zeta)
    #damping frequency omeg_d * t calculation
    omega_d_t=omega_t*np.sqrt(1-zeta**2)
    #c(wt) calculation
    c_omega_t=1-(1/np.sqrt(1-zeta**2))*np.exp(-zeta*omega_t)*np.sin(omega_d_t+angle)
    ax.plot(omega_t, c_omega_t, label=fr"$\zeta={np.around(zeta,2)}$")
    plt.legend()
ax.set_xlabel("$\omega_n t$", fontsize=20)
ax.set_ylabel("$c(\omega_n t)$", fontsize=20)
ax.set_xlim(0, 17)
ax.set_ylim(0,2)
ax.grid(True)
```

### Second order system response

```python
#second-order system response for different zeta and omega
def system_response(zeta, omega):
    s=co.tf("s")
    G=omega**2/(s**2+2*zeta*omega*s+omega**2)
    C=G*(1/s) #system output for input step function
    fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,7))
    #pole-zero plot
    poles=np.around(C.pole())
    ax[0].plot(poles.real, poles.imag, marker="x",
              markersize=15, 
              color="blue", 
              linestyle="none")
    #move axis to center
    def spine_origin(axes_index, xlabel, ylabel, title):
        #move left y axis and bottom x axis to center (0,0)
        ax[axes_index].spines["left"].set_position(("data",0))
        ax[axes_index].spines["bottom"].set_position(("data", 0))
        #Eliminate upper and right axes
        ax[axes_index].spines["right"].set_visible(False)
        ax[axes_index].spines["top"].set_visible(False)
        ax[axes_index].set_xlabel(xlabel)
        ax[axes_index].set_ylabel(ylabel)
        ax[axes_index].set_title(title, fontdict=font)
    spine_origin(0, None, None, "Poles plot on S plane")
    ax[0].grid()
    #step response plot
    t, y=co.step_response(G)
    ax[1].plot(t, y, color="blue")
    ax[1].grid()
    spine_origin(1, "time", "amplitude", 
                 "Second-order system step response")
# defint text font properties
font = {'family': 'serif',
        'color':  'DarkGreen',
        'weight': 'normal',
        'size': 18,
        }
interact(system_response, zeta=(0, 2, 0.1), omega=(6,10,1))
```

### Evaluation of %OS

### System response with additional pole

```python
def additional_pole(p):
    s=co.tf("s")
    #define transfer function
    G=24.542/(s**2+4*s+24.542)
    Gp=(24.542)/(((1/p)*s+1)*(s**2+4*s+24.542))
    #step response of second and thire order system
    t=np.arange(0, 6, 0.01)
    _, yp=co.step_response(Gp, t)
    _, y=co.step_response(G, t)
    fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
    #determine pole of a system
    Cp=Gp*(1/s)

    pole_p=Cp.pole()
    #pole plot
    ax[0].plot(pole_p.real, pole_p.imag, 
              marker="x", 
              markersize=15, 
              linestyle="none", 
              color="DarkBlue")
    #step response plot
    ax[1].plot(t, yp, 
           linewidth=2, 
           color="DarkBlue", 
           label="3rd order")
    ax[1].plot(t, y, 
           linewidth=2, 
           color="Red", 
           label="2nd order")
    
    #move axis to center
    def spine_origin(axes_index, xlabel, ylabel, title):
        #move left y axis and bottom x axis to center (0,0)
        ax[axes_index].spines["left"].set_position(("data",0))
        ax[axes_index].spines["bottom"].set_position(("data", 0))
        #Eliminate upper and right axes
        ax[axes_index].spines["right"].set_visible(False)
        ax[axes_index].spines["top"].set_visible(False)
        ax[axes_index].set_xlabel(xlabel)
        ax[axes_index].set_ylabel(ylabel)
        ax[axes_index].set_title(title, fontdict=font)
    
    spine_origin(0, None, None, "Poles plot on S plane")
    spine_origin(1, "Time(s)", "Amplitude", "2nd & 3rd order system")
    ax[0].set_xlim(-p-1, 0)
    plt.legend()
    ax[0].grid()
    ax[1].grid()
interact(additional_pole, p=(0.001, 30, 1))
```

### System response with LHP zero

```python
def additional_zero(z):
    s=co.tf("s")
    #define transfer function
    C=24.542/(s**2+4*s+24.542)
    C_z=((1/z)*s+1)*24.542/(s**2+4*s+24.542)
    #step response of second and thire order system
    t=np.arange(0, 4, 0.01)
    _, c_z=co.step_response(C_z, t)
    _, c=co.step_response(C, t)
    fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
    #determine pole of a system
    pole=C.pole()
    C_z_pole=C_z.pole()
    C_z_zero=C_z.zero()
    #pole plot
    ax[0].plot(C_z_pole.real, C_z_pole.imag, 
              marker="x", 
              markersize=15, 
              linestyle="none", 
              color="red")
    #zero plot
    ax[0].plot(C_z_zero.real, C_z_zero.imag, 
              marker="o", 
              markersize=15,
              markerfacecolor="white", 
              color="red")
    #step response plot
    ax[1].plot(t, c_z, 
           linewidth=2, 
           color="DarkBlue", 
           label="With Zero")
    ax[1].plot(t, c, 
           linewidth=2, 
           color="Red", 
           label="Without Zero")
    
    #move axis to center
    def spine_origin(axes_index, xlabel, ylabel, title):
        #move left y axis and bottom x axis to center (0,0)
        ax[axes_index].spines["left"].set_position(("data",0))
        ax[axes_index].spines["bottom"].set_position(("data", 0))
        #Eliminate upper and right axes
        ax[axes_index].spines["right"].set_visible(False)
        ax[axes_index].spines["top"].set_visible(False)
        ax[axes_index].set_xlabel(xlabel)
        ax[axes_index].set_ylabel(ylabel)
        ax[axes_index].set_title(title, fontdict=font)
    
    spine_origin(0, None, None, "Pole-Zero plot on S plane")
    spine_origin(1, "Time(s)", "Amplitude", 
                 "2nd order system With LHP zero")
    ax[0].set_xlim(-z-1, 0)
    plt.legend()
    ax[0].grid()
    ax[1].grid()
interact(additional_zero, z=(1, 20, 1))
```
